```python
import matplotlib.pyplot as plt

class Election:
    def __init__(self, vote_shares, total_seats, colors):
        self.vote_shares = vote_shares
        self.total_seats = total_seats
        self.colors = colors
        self.seats = {}

    def dhont(self):
        # Initialize seats to be zero for all parties
        seats = {party: 0 for party in self.vote_shares.keys()}
        # Initialize scores to be vote shares for all parties
        scores = self.vote_shares.copy()

        for _ in range(self.total_seats):
            # Find the party with the highest score
            max_party = max(scores, key=scores.get)

            # Allocate a seat to this party
            seats[max_party] += 1

            # Update the score for this party
            scores[max_party] = self.vote_shares[max_party] / (seats[max_party] + 1)
        
        self.seats = seats

    def plot_seats(self):
        # Get party names and corresponding seat counts
        parties = list(self.seats.keys())
        seat_counts = list(self.seats.values())

        # Create a bar chart
        bar_plot = plt.bar(parties, seat_counts, color=[self.colors[party] for party in parties])

        # Add labels
        plt.xlabel('Party')
        plt.ylabel('Number of Seats')
        plt.title('Seat Distribution')

        # Tilt the x-axis labels
        plt.xticks(rotation=45)

        # Add seat count labels on top of each bar
        for bar in bar_plot:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, yval, ha='center', va='bottom')

        # Display the plot
        plt.show()

    def compute_and_plot(self):
        self.dhont()
        self.plot_seats()
```

```python
vote_shares = {"Partido Colorado": 1288722, "Partido Liberal": 685694, "Partido de Payo":323580, "Partido Encuentro":144977, "Partido Otros": (2947305-1288722-685694-323580-144977)}
total_seats = 45
colors = {"Partido Colorado": "red", "Partido Liberal": "blue", "Partido de Payo": "gray", "Partido Encuentro":"yellow", "Partido Otros": "green"}

election = Election(vote_shares, total_seats, colors)
election.compute_and_plot()
```
